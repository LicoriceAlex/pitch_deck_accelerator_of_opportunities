### Making style

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
Image.open('hh.png')
```

```python
def dominant_colors(img_path, n_dom_colors):
    img = Image.open(img_path)
    img = img.quantize(colors=n_dom_colors,
                       kmeans=n_dom_colors).convert('RGB')

    dom_colors = sorted(img.getcolors(2 ** 24), reverse=True)[:n_dom_colors]
    
    return [i[1] for i in dom_colors]
```

```python
palette = np.array(dominant_colors('hh.png', 4))[np.newaxis, :, :]

plt.imshow(palette)
plt.axis('off')
plt.show()
```

```python
s = palette[0].sum(axis=1)

darkest_color   = palette[0][s.argmin()]
brightest_color = palette[0][s.argmax()]

darkest_color, brightest_color
```
